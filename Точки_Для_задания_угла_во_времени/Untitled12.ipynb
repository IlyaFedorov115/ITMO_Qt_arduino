```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python

with open('set_points.txt', 'r') as file:
    # Читаем содержимое файла с помощью csv.reader
    reader = csv.reader(file)

    # Инициализируем два массива для хранения значений из столбцов
    column1 = []
    column2 = []

    # Проходим по каждой строке в файле
    for row in reader:
        # Если строка пустая, пропускаем её
        if not row:
            continue

        # Добавляем элементы из первого и второго столбца в соответствующие массивы
        #column1.append(row[0])
        #column2.append(row[1])
        column1.append(float(row[0]))  # преобразуем в числа
        column2.append(float(row[1]))  # преобразуем в числа

# Выводим значения из каждого массива через пробел
#print(' '.join(column1))
#print(' '.join(column2))
```

```python
column1 = np.array(column1)
column2 = np.array(column2)
column1 -= 55
```

```python
for e in column2:
  print('%.3f' % e, end=', ')
print("\n" + str(len(column1)))
```

```python
plt.plot(column1, column2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('График из файла CSV')
#plt.grid(True)
plt.show()
```

```python

from scipy.interpolate import interp1d
# Преобразуем массивы в numpy массивы для удобства работы с ними
x_values = np.array(column1)
y_values = np.array(column2)

# Выполняем интерполяцию
interpolation_function = interp1d(x_values, y_values, kind='linear')

# Создаем новый массив значений для x
x_new = np.linspace(x_values.min(), x_values.max(), 1000)

# Получаем соответствующие значения для y с помощью интерполяционной функции
y_new = interpolation_function(x_new)

# Строим график
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'o', label='Исходные данные')
plt.plot(x_new, y_new, '-', label='Интерполяция')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Интерполяция данных из файла CSV')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def func(x, a1, b1, c, a2, b2, a3, b3, a4, b4, a5, b5, a6, b6, t1, t2, t3, t4, t5):
    return a1 * np.sin(b1 * x + t1) + c + a2 * np.sin(b2 * x + t2) + a3 * np.cos(b3 * x + t3) + a4 * np.cos(b4 * x + t4) + a5 * np.sin(b5 * x + t5) + a6 * np.sin(b6 * x)
```

```python
# Выполняем аппроксимацию данных
x_data = x_new
y_data = y_new
popt, pcov = curve_fit(func, x_data, y_data)

# Получаем коэффициенты аппроксимирующей функции
#a1, b1, c, a2, b2, a3, b3 = popt

# Выводим полученные коэффициенты
#print("a1 =", a1)
#print("b1 =", b1)
#print("c =", c)
#print("a2 =", a2)
#print("b2 =", b2)
#print("a3 =", a3)
#print("b3 =", b3)
print(list(popt))

# Строим график исходных данных и аппроксимирующей функции
plt.scatter(x_data, y_data, label='Исходные данные')
plt.plot(x_data, func(x_data, *popt), 'r-', label='Аппроксимация: ')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Аппроксимация данных из файла CSV')
plt.legend()
plt.grid(True)
plt.show()
```
